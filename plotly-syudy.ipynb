# Plotly Express
인터랙티브 시각화 고수준 API 입니다. Pandas DF로 바로 연결
(저수준 API도 있어요. 세밀한 제어가 가능.)

```python
import plotly.express as px
import pandas as pd
```

```python
# 데이터 불러오기
imdb = pd.read_csv('./imdb_top_1000.csv')

# 데이터 크기 확인
imdb.shape 
```

```python
imdb.head(3)
```

```python
# 결측치 확인
imdb.isnull().sum()
```

```python
# 데이터가 잡음(노이즈) 규격에 맞지 않는 데이터가 생깁니다.
imdb['Released_Year'].value_counts().tail(10)
```

## 실제 데이터는 여러가지 문제를 갖는다.
결측치, 노이즈, 구분된 문자열 쪼개기 등..

```python
fig = px.histogram(
    imdb,
    x='IMDB_Rating',
    nbins=15,
    marginal='box',     # 박스플롯 동시에!
    title='IMDB 평점 분포',
    template='simple_white'
)
fig.show()
```

```python
fig = px.scatter(
    imdb,
    x='No_of_Votes',
    y='IMDB_Rating',
    hover_name='Series_Title',      # 굵게 표시
    hover_data=['Released_Year',
                'Genre'],
    title='투표 수 vs IMDB 평점',
    template='simple_white',
    opacity=0.6
)
fig.show()
fig.write_html('imdb_scatter.html')
```

```python
# 막대차트 그리기
df = imdb.dropna(subset=['Released_Year']).copy() # 결측치 제거해서 사본 만들기
df['Released_Year'] = pd.to_numeric(df['Released_Year'], errors='coerce')
df = df.dropna(subset=['Released_Year'])
df['Released_Year'] = df['Released_Year'].astype('int')
year_count = df['Released_Year'].value_counts().sort_index()
# 전처리는 빡셈
type(year_count)
```

```python
# 시각화 코드는 단순
fig = px.bar( x=year_count.index, y=year_count.values, title='연도별 상위 영화 수' )
fig.show()
```
